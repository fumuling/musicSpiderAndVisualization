```python
from pyecharts.charts import Line, Bar, Pie
from pyecharts.globals import CurrentConfig, NotebookType
CurrentConfig.NOTEBOOK_TYPE = NotebookType.JUPYTER_LAB
from pyecharts import options as opts
import pandas as pd
import numpy as np
from wordcloud import WordCloud
import matplotlib.pyplot as plt
```

```python
res = pd.read_csv('wordsFre.csv', encoding='gbk')
res.head(10)
```

```python
a = np.array(res.head(10))
a.tolist()
```

```python
datax = [i[0] for i in a.tolist()]
datay = [i[1] for i in a.tolist()]
bar = (
    Bar()
    .add_xaxis(datax)
    .add_yaxis("词频", datay)
    .set_global_opts(title_opts=opts.TitleOpts(title="最近到底在听啥"), yaxis_opts = opts.AxisOpts())
)
bar.load_javascript()
```

```python
bar.render_notebook()
```

```python
singer_dict = {}
with open('singer.txt', 'r') as f:
    for line in f.readlines():
        if line[:-1] in singer_dict:
            singer_dict[line[:-1]] += 1
            
        else:
            singer_dict[line[:-1]] = 1
singer_dict
```

```python
foo = ' '.join(list(singer_dict.keys()))
foo
```

```python
wc = WordCloud(
    font_path=font,
    background_color='white',
    max_font_size=100,
    width=500,
    height=350,
    min_font_size=5,
    mode='RGBA'
)
wc.generate(foo)
```

```python
plt.imshow(wc, interpolation='bilinear')
plt.axis("off")
plt.show()
```

```python
pie = (
    Pie()
    .add('', [list(z) for z in singer_dict.items()])
    .set_global_opts(title_opts = opts.TitleOpts(title = ''),
                    legend_opts = opts.LegendOpts(is_show = False))
    .set_series_opts(label_opts = opts.LabelOpts(formatter = '{b}:{d}%'))
)
```

```python
pie.render_notebook()
```

```python
name_list = []
with open('songName.txt', 'r',encoding='utf-8') as f:
    for i in f.readlines():
        name_list.append(i.strip())
len(name_list)
name_list[:10]
```

```python
count_list = []
with open('count.txt', 'r') as f:
    for i in f.readlines():
        count_list.append(i.strip()[:-1])
count_list[:10]
```

```python
list1 = zip(name_list, count_list)
list1
```

```python
datax = [i for i in name_list[:10]]
datay = [i for i in count_list[:10]]
bar = (
    Bar()
    .add_xaxis(datax)
    .add_yaxis("听歌次数", datay)
    .set_global_opts(title_opts=opts.TitleOpts(title="最近到底在听啥"), yaxis_opts = opts.AxisOpts())
    .reversal_axis()
)
```

```python
bar.render_notebook()
```
